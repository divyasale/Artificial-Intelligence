### Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

##### Basics of plotting

```python
#Line Plot

x=[1,2,3]
y=[2,4,1]

plt.plot(x,y)

plt.title("Simple Line plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid()
plt.show()    #Function to show the plot
```

```python
## Plot sine wave using numpy and matplotlib

x=np.arange(0,10,0.1)
y=np.sin(x)

print(x)

print(y)
```

```python
print(y.min())
```

```python
print(y.max())
```

```python
#Sine Wave
plt.plot(x,y)

plt.title("Sine Wave")
plt.xlabel("X Value")
plt.ylabel("Sin(x) values")
plt.show()
```

```python
#Cos Wave
x=np.arange(0,10,0.1)
y=np.cos(x)
plt.plot(x,y)

plt.title("Cos Wave")
plt.xlabel("X Value")
plt.ylabel("Cos(x) values")
plt.show()
```

```python
#tan Wave
x=np.arange(0,10,0.1)
y=np.tan(x)
plt.plot(x,y)

plt.title("Tan Wave")
plt.xlabel("X Value")
plt.ylabel("Tan(x) values")
plt.show()
```

```python
#Cot wave
x=np.arange(0,10,0.1)
y=1/np.tan(x)
plt.plot(x,y)

plt.title("Cot Wave")
plt.xlabel("X Value")
plt.ylabel("Cot(x) values")
plt.show()
```

```python
#Sec Wave
x=np.arange(0,10,0.1)
y=1/np.cos(x)
plt.plot(x,y)

plt.title("Sec Wave")
plt.xlabel("X Value")
plt.ylabel("Sec(x) values")
plt.show()
```

```python
#Cosec Wave
x=np.arange(0,10,0.1)
y=1/np.sin(x)

plt.plot(x,y)
plt.title("Cosec Wave")
plt.xlabel("X Value")
plt.ylabel("Cosec(x) values")
plt.show()
```

```python
a=np.arange(0,10,0.1)
b=np.sin(a)
c=np.cos(a)
d=np.tan(a)
e=1/np.tan(a)


plt.subplot(2,2,1)
plt.plot(a,b)
plt.figure(figsize=(6,3))
plt.title("Sin wave")

plt.subplot(2,2,2)
plt.plot(a,c)
plt.figure(figsize=(6,3))
plt.title("Cos wave")

plt.subplot(2,2,3)
plt.plot(a,d)
plt.title("Tan wave")

plt.subplot(2,2,4)
plt.plot(a,e)
plt.title("Cot wave")
```

##### Some more examples of line plot

```python
# Let us consider 4 list, days, max temp, min temp and avg temp
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label='max_temp',color="red")
plt.plot(days,min_t,label='min_temp',color="magenta")
plt.plot(days,avg_t,label='avg_temp',color='blue')
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Days Vs temp")
plt.legend(loc="best")

plt.show()
```

## Text and Annotations

```python
## Scatter Plot sine wave using numpy and matplotlib

fig,ax=plt.subplots()
x=np.arange(0,10,0.1)
y=np.sin(x)

ax.plot(x,y)
ax.annotate("local maxima",xy=(1.6,1),xytext=(4,0.75),arrowprops=dict(facecolor="Orange"))
ax.annotate("local minima",xy=(4.6,-1),xytext=(7,-0.85),arrowprops=dict(facecolor="Orange"))
ax.annotate("next maxima",xy=(8,1))
```

### 3D Plotting

```python
from mpl_toolkits import mplot3d
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Empty 3D Graph

%matplotlib notebook

fig=plt.figure()    #Create figure
ax=plt.axes(projection='3d')  #syntax for 3d plotting
```

```python
#3D Diagram  - 3D line plot

x=np.linspace(0,10,1000)
y=np.sin(x)
z=np.cos(x)

fig=plt.figure()   
ax=plt.axes(projection='3d')
ax.plot3D(x,y,z)
```

```python
#Another Example
x=np.linspace(0,10,1000)
y=x**2
z=x**3

fig=plt.figure()   
ax=plt.axes(projection='3d')
ax.plot3D(x,y,z)
```

```python
#3D Line plot
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

fig=plt.figure()   
ax=plt.axes(projection='3d')
ax.plot3D(x,y,z)
```

```python
#3D scatter plot
x=np.linspace(0,15,1000)
y=np.sin(x)
z=np.cos(x)

fig=plt.figure()   
ax=plt.axes(projection='3d')
ax.scatter3D(x,y,z)
```
